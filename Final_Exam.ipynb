manual rewrite with doctest:

```python
# Section 2.2 - Corrected Task Manager Code
# Course constraint: Week 5 (Lists and Loops), Week 12 (Testing and Debugging)

tasks = []

def add_task(task):
    """
    Adds a task to the task list.

    >>> add_task("Wash dishes")
    Added: Wash dishes
    """
    tasks.append(task)
    print(f"Added: {task}")

def remove_task(index):
    """
    Removes a task from the task list by index.

    >>> add_task("Finish exam")
    Added: Finish exam
    >>> remove_task(0)
    Removed: Finish exam
    """
    if index < len(tasks):
        removed = tasks[index]
        del tasks[index]
        print(f"Removed: {removed}")
    else:
        print("Invalid index!")

def list_tasks():
    """
    Prints all tasks in the list.

    >>> list_tasks()  # doctest: +SKIP
    """
    if not tasks:
        print("No tasks available.")
    else:
        for i, t in enumerate(tasks):
            print(f"{i+1}: {t}")

def main():
    add_task("Buy milk")
    add_task("Pay bills")
    add_task("Walk dog")
    list_tasks()
    remove_task(1)
    list_tasks()

if __name__ == "__main__":
    import doctest
    doctest.testmod(verbose=True)
```

```python
# Section 3.2 - Refined WeatherWise Function
# Course constraint: Week 8 – APIs, JSON, and Basic Error Handling

def refined_safe_weather_data_fetch(city):
    """
    Fetch weather data with basic error handling - Week 8 style

    >>> refined_safe_weather_data_fetch("")
    Error: City name cannot be empty
    >>> refined_safe_weather_data_fetch("Perth")  # doctest: +SKIP
    {'city': 'Perth', 'temperature': '25', 'wind_speed': '13', 'description': 'Partly cloudy'}
    """
    if not city:
        print("Error: City name cannot be empty")
        return None

    try:
        import requests
        url = f"http://wttr.in/{city}?format=j1"
        response = requests.get(url)
        data = response.json()

        current = data["current_condition"][0]
        description = current["weatherDesc"][0]["value"]
        temperature = current["temp_C"]
        wind_speed = current["windspeedKmph"]

        return {
            "city": city,
            "temperature": temperature,
            "wind_speed": wind_speed,
            "description": description
        }
    except:
        print("Error: Could not retrieve weather data.")
        return None
```

4.1 Manual Implementation with Course Constraints

```python
# WEEK 3 – Simple static menu printing using print()
def display_menu():
    """
    Display menu options - Week 3 formatting only

    >>> display_menu()  # doctest: +SKIP
    1. Add task
    2. List tasks
    3. Remove task
    4. Quit
    """
    print("1. Add task")
    print("2. List tasks")
    print("3. Remove task")
    print("4. Quit")

    # WEEK 6 – Input validation using basic if/else and try/except
def get_user_choice():
    """
    Get user input and validate between 1 and 4 (Week 6 style)

    >>> # Manual test only
    >>> # get_user_choice()  # doctest: +SKIP
    """
    choice = input("Enter your choice (1-4): ")
    if choice.isdigit():
        num = int(choice)
        if 1 <= num <= 4:
            return num
    print("Invalid input. Please enter a number from 1 to 4.")
    return None
    # WEEK 5 – Lists + loops only. No complex logic.
def main():
    """Main programme loop using Week 5 patterns only"""
    tasks = []

    while True:
        display_menu()
        choice = get_user_choice()

        if choice is None:
            continue

        if choice == 1:
            task = input("Enter a new task: ")
            tasks.append(task)
            print("Task added.")

        elif choice == 2:
            if len(tasks) == 0:
                print("No tasks available.")
            else:
                for i in range(len(tasks)):
                    print(str(i) + ": " + tasks[i])

        elif choice == 3:
            if len(tasks) == 0:
                print("No tasks to remove.")
            else:
                index = input("Enter task index to remove: ")
                if index.isdigit():
                    index = int(index)
                    if 0 <= index < len(tasks):
                        removed = tasks.pop(index)
                        print("Removed task:", removed)
                    else:
                        print("Invalid index.")
                else:
                    print("Please enter a number.")

        elif choice == 4:
            print("Exiting... Goodbye!")
            break
```
